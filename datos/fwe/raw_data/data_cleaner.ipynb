```python
import pandas as pd
import os
import openpyxl
```

```python
#Get the base directory (where the script is located)
base_dir = os.getcwd()
# List all .xlsx files in the base directory
excel_files = [f for f in os.listdir(base_dir) if f.endswith('.xlsx')]
# Read each .xlsx file into a DataFrame
dfs = [pd.read_excel(os.path.join(base_dir, f)) for f in excel_files]
# Concatenate all DataFrames into one
df = pd.concat(dfs, ignore_index=True)
# Show the result
df.head()
```

```python
df['Food Product'].unique()
```

```python
df_cleaned=df[df['Compound Group']=='Proximates'].reset_index(drop=True)
df_cleaned.head()
```

```python
df_cleaned.shape
```

```python
df_cleaned.to_excel("Combined_foods.xlsx", index=False)
```

```python

combined_foods = pd.read_excel(os.path.join(base_dir, 'Combined_foods.xlsx'))
todo_filtrado = pd.read_excel(os.path.join(base_dir, 'todo_filtrado.xlsx'))
# Concatenate all DataFrames into one
df_final = pd.concat([combined_foods,todo_filtrado], ignore_index=True).sort_values(by='Food Product').reset_index(drop=True)
# Show the result
df_final.head()
```

```python
df_final.to_excel("Combined_foods_complete.xlsx", index=False)
```
